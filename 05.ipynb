```python
from datasets import load_dataset
```

```python
texts = load_dataset("csv", data_files="./data/redacoes.csv")
```

```python
texts["train"]
```

```python
texts = texts["train"].train_test_split(test_size=0.2, shuffle=False)
texts
```

```python
model_checkpoint = "Geotrend/distilbert-base-pt-cased"
```

```python
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModel.from_pretrained(model_checkpoint)
```

```python
tokenizer.tokenize(
    texts["train"]["essay"][0]
)
```

```python
tokenizer.convert_tokens_to_ids(
    tokenizer.tokenize(
        texts["train"]["essay"][0]
    )
)
```

```python
tokenizer.decode(
    tokenizer.convert_tokens_to_ids(
        tokenizer.tokenize(
            texts["train"]["essay"][0]
        )
    )
)
```

```python
tokenizer(
    texts["train"]["essay"][0],
    return_tensors="pt"
)
```

```python
def tokenize_fn(text_data):
    return tokenizer(text_data["essay"], truncation=True)
```

```python
tokenized_texts = texts.map(tokenize_fn, batched=True, remove_columns=["essay"])
```

```python
tokenized_texts
```

```python
tokenized_texts["train"].to_pandas()
```

```python
tokenized_texts["train"].features["score"]
```

```python
from datasets import ClassLabel
```

```python
scores = ClassLabel(names=[str(i) for i in range(11)])

scores
```

```python
def label_map_fn(data):
    data["label"] = scores.str2int(str(data["score"]))

    return data
```

```python
tokenized_texts = tokenized_texts.map(label_map_fn, remove_columns=["score"])
```

```python
tokenized_texts
```

```python
tokenized_texts = tokenized_texts.cast_column("label", scores)
```
